# Part III: Configure street width pixels (Bölüm III: Sokak genişliği piksellerini yapılandırma)

OSMnx, sokak genişliklerini piksel cinsinden belirtmek için otoyol OSM etiketini kullanır. Varsayılan değerleri geçersiz kılmak için bir sokak genişlikleri sözlüğüne geçebilirsiniz. Sokak genişlikleri diktesinde anahtar olarak görünmeyen herhangi bir sokak türü, default_width değerine geri döner.

```python
street_widths = {
    "footway": 0.5,
    "steps": 0.5,
    "pedestrian": 0.5,
    "path": 0.5,
    "track": 0.5,
    "service": 2,
    "residential": 3,
    "primary": 5,
    "motorway": 6,
}
place = "sf-custom"
point = (37.793897, -122.402189)
fp = f"./{img_folder}/{place}.{extension}"
fig, ax = ox.plot_figure_ground(
    point=point,
    filepath=fp,
    network_type="all",
    street_widths=street_widths,
    dpi=dpi,
    save=True,
    show=False,
    close=True,
)
Image(fp, height=size, width=size)
```
